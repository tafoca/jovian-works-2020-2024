```python
#import required libraries
from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree 
from sklearn.preprocessing import LabelEncoder 
from sklearn.tree import export_graphviz
from graphviz import Source
from IPython.display import SVG
import pandas as pd 
import numpy as np 
import io
import requests
```

```python
#Read training Dataset 
url="https://raw.githubusercontent.com/tafoca/Datasets/main/student-data.csv"
df=pd.read_csv(url)
print(df)
```

```python
# Prepare your dataset 
lb = LabelEncoder() 
df['PSM_'] = lb.fit_transform(df['PSM']) 
df['CTG_'] = lb.fit_transform(df['CTG'] ) 
df['SEM_'] = lb.fit_transform(df['SEM'] )  
df['ASSS_'] = lb.fit_transform(df['ASSS'] ) 
df['ATT_'] = lb.fit_transform(df['ATT'] ) 
df['LW_'] = lb.fit_transform(df['LW'] ) 
df['ESM_'] = lb.fit_transform(df['ESM'] ) 

X = df.iloc[:,7:13] 
Y = df.iloc[:,13]
```

```python
df
```

```python
#X,Y
```

```python
print("Features in numeric form")
print(X)
print("*********************************************")
print("Label in numeric form")
print(Y)
# Prepare a testcase
X_test=[[0 for x in range(6)] for y in range(1)]

X_test
```

```python
X_test[0][0]=1
X_test[0][1]=1
X_test[0][2]=1
X_test[0][3]=0
X_test[0][4]=1
X_test[0][5]=1
#Train a Decision tree
d_tree = DecisionTreeClassifier(criterion='entropy')
d_tree.fit(X.astype(int), Y.astype(int))
#Test using sample test X_test
y_pred_en = d_tree.predict(X_test)
#The prediction for X-test
print(y_pred_en)
```

```python
#Visualize the decision tree
graph = Source(export_graphviz(d_tree, out_file=None, feature_names=X.columns))
SVG(graph.pipe(format='svg'))
```
