```python
# importing the pandas library
import pandas as pd

# Reading the csv file and storing it
# in a variable
inputfile = "ue_583-589_MED3_2022.csv";
ouputfile = "ue_583-589_MED3_2022-bin.csv";
df = pd.read_csv(inputfile)
```

```python
df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']] = df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']].replace(1, 0)
df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']] = df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']].replace([2,3,4], -1)
df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']] = df.loc[:,['c','d','e','f','g','h', 'i', 'j', 'k', 'l', 'm', 'n','o']].replace([5,6,7], 1)
```
